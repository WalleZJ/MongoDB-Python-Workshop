```python
for r in restaurants.find({'cuisine': 'Delicatessen'}, {'_id':0, 'borough': 1}):
    print(r)
```

```python
restaurants.distinct('cuisine')
```

```python
regx = re.compile("^deli", re.IGNORECASE)
```

```python
for r in restaurants.find({'name': {'$regex': regx}}):
    print(r)
```
